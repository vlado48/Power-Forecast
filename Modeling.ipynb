## Time-series forcasting of power

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.api.types import is_numeric_dtype
from hyperopt import fmin, tpe, hp, anneal, Trials, pyll

data = pd.read_csv('dataset_tidy.csv')
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data = data.set_index('Timestamp')
data = data['power']
```

#### Data splits

```python
from statsmodels.tsa.seasonal import STL

minutes_day = 1440
data_array = data.to_numpy().T
data_array = data_array[minutes_day:]    # Skip first day

# Data split into train, validation and test
l = len(data_array)
X_train, X_valid, X_test = np.split(data_array, [int(l*0.7), int(l*0.9)])

print(f'Train size: {len(X_train)}, Validation size: {len(X_valid)}, Test size: {len(X_test)}')
```

## ARIMA

```python
from statsmodels.tools.eval_measures import meanabs

def eval(trained, X_valid, metric='rmse'):
    '''evaluates models prediction and returns either rmse or mape metric'''
    forecast = np.zeros(len(X_valid))       # Declare empty array to hold forecasts   
    for i in range(len(X_valid)):
        forecast[i] = trained.forecast()    # Append forcasted minute
        trained     = trained.extend(([X_valid[i]]))    # Append actual value, retrain and repeat loop

    if metric == 'rmse':
        resid = forecast - X_valid     
        rmse  = np.sqrt(np.sum(resid**2) / len(X_valid))
        print('Model RMSE:', rmse)    
        return rmse
    
    elif metric == 'mape':
        mape = ((X_valid - forecast) / X_valid).mean()
        print('Model MAPE:', mape)         
        return mape

def hparam_eval(params, X_train=X_train, X_valid=X_valid, metric='rmse'):
    '''evaluates model using specific hyperparameters, used for optimization'''
    model = sm.tsa.SARIMAX(X_train, **params)
    print('Model initialized with ', params)
    trained = model.fit()
    print('Model trained')    
    return eval(trained, X_valid, metric=metric)
```

## ARIMAX
- uses 1m based, smoothed daily average for exogenious regression 
- !!! Fundamental mistake, given daily average was constructed using the future data -> target leakage

```python
# Open daily smoothed average
exog = pd.read_csv('power_D_avg.csv')
exog = exog['0'].to_numpy().T

# Expande across entire time series
l = len(data_array)
while len(exog)<l:
    exog = np.append(exog, exog)
exog = exog[:l]

# exogenius time-series split
exog_train, exog_valid, exog_test = np.split(exog, [int(l*0.7), int(l*0.9)])

print(f'Exog Train size: {len(exog_train)},\
      \nExog Validation size: {len(exog_valid)},\
      \nExog Test size: {len(exog_test)}')
```

```python
# Parameters space for ARIMAX 
param_space_exo = {'order': ((hp.choice('AR', [0,1,2,3])),
                             (hp.choice('I',  [0,1,2])),
                             (hp.choice('MA', [0,1,2,3]))),
                   'trend': hp.choice('trnd', ['n','c','t','ct'])
                  }
```

```python
def eval_exo(trained_full, X_train, X_valid, metric='rmse', plot=False):
    '''evaluates models prediction and returns either rmse or mape metric'''
    predictions = trained_full.predict(start=len(X_train))
    resid = predictions - X_valid
    
    if plot:
        plt.figure(figsize=(13,13))
        plt.plot(X_valid)
        plt.plot(predictions)
        plt.title('Predictions')
        plt.legend(['Data', 'Predictions'])
        plt.show()
    
    if metric == 'rmse':
        rmse  = np.sqrt(np.sum(resid**2) / len(X_valid))
        print('Model RMSE:', rmse)    
        return rmse    
    elif metric == 'mape':
        mape = ((X_valid - predictions) / X_valid).mean()
        print('Model MAPE:', mape)          
        return mape
    
def hparam_eval_exo(params, X_train=X_train, X_valid=X_valid, exog_train=exog_train, exog_valid=exog_valid, metric='rmse', plot=False):
    '''evaluates model using specific hyperparameters, used for optimization'''
    params['exog'] = exog_train
    print('Model initialized with ', params)        
    model = sm.tsa.SARIMAX(X_train, **params)             
    trained = model.fit(disp=False)
    print('Model trained')    
     
    # Training + validation SARIMAX (prediction)   
    params['exog'] = np.append(exog_train, exog_valid)   
    model = sm.tsa.SARIMAX(np.append(X_train, X_valid), **params)   
    trained_full = model.filter(trained.params)    # Use the parameters from training model
    
    return eval_exo(trained_full, X_train, X_valid, metric=metric, plot=plot)
```

```python
trials_exo = Trials()
best_exo   = fmin(fn=hparam_eval_exo,  # function to optimize
              space=param_space_exo, 
              algo=tpe.suggest,        # optimization algorithm, hyperotp will select its parameters automatically
              max_evals=192,           # maximum number of iterations
              trials=trials_exo)       # logging
```

## Evaluation & Comparison

Naive model as baseline - it expect next value to be same as previous value

```python
naive_rmse = np.sum(np.diff(X_test)**2) / (len(X_test)-1)
naive_mape = np.abs((np.diff(X_test) / X_test[1:])).mean()
```

ARIMAX rmse and mape

ARIMA rmse and mape

```python
arima_rmse = hparam_eval({'order': (0,2,0), 'trend': 'n'}, 
                         X_train= np.append(X_train, X_valid),
                         X_valid= X_test)
arima_mape = hparam_eval({'order': (0,2,0), 'trend': 'n'}, 
                         X_train= np.append(X_train, X_valid),
                         X_valid= X_test,
                         metric='mape')
```
